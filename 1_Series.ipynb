# Exercise 1 • Test scores

Create a series of 10 elements, random integers from 70 to 100, representing scores on a monthly exam. Set the index to be the month names, starting in September and ending in June. (If these months don’t match the school year in your location, feel free to make them more realistic.)

With this series, write code to answer the following questions:

1. What is the student’s average test score for the entire year?
1. What is the student’s average test score during the first half of the year (i.e., the first five months)?
1. What is the student’s average test score during the second half of the year?
1. Did the student improve their performance in the second half? If so, by how much?

```python
from pandas import Series, DataFrame
import pandas as pd
from numpy.random import randn
import numpy as np
```

```python
#seeds random with 0 to make results the same as the book
g = np.random.default_rng(0)
months = 'Sep Oct Nov Dec Jan Feb Mar Apr May Jun'.split()
s = Series(g.integers(70,101,10), index=months)
s
```

```python
#1. What is the student’s average test score for the entire year?
s.mean()
```

```python
#2. What is the student’s average test score during the first half of the year (i.e., the first five months)?
s[:5].mean()
```

```python
#3. What is the student’s average test score during the second half of the year?
s[5:].mean()
```

```python
#4. Did the student improve their performance in the second half? If so, by how much?
first_half_average = s.loc['Sep':'Jan'].mean()
second_half_average = s.loc['Feb':'Jun'].mean()
diff = second_half_average - first_half_average
print(f'First half average: {first_half_average}')
print(f'Second half average: {second_half_average}')
print(f'Improvement: {"up" if diff > 0 else "down"} {abs(diff):.2f}')
```

```python
#5. In which month did this student get their highest score?
#Note that there are at least three ways to accomplish this: you can sort the values, taking the largest one, using a boolean (mask) index to find rows that match the value of s.max(),
#the highest value, or invoking s.idxmax(), which returns the index of the highest value.
m1 = s[s.idxmax()]
m2 = s.max()
m3 = s.sort_values(ascending=False).head(1).iloc[0]
if m1 != m2 or m1 != m3:
  raise ValueError('Values must be the same!')
print(f'Highest score: {m3}')
```

```python
#6. What were this student’s five highest scores?
for label, score in enumerate(s.sort_values(ascending=False).head(5)):
  print(f'{label+1}. {score}')
s.sort_values(ascending=False).head(5)
```

```python
#7. Round the student’s scores to the nearest 10. (A score of 82 would be rounded down to 80, but a score of 87 would be rounded up to 90.)
#Be sure to read the documentation for the round method (http://mng.bz/8rzg) to understand its arguments and how it handles numbers like 15 and 75.
df_round = np.round(s/10)*10
df_round
```

#Exercise 2 • Scaling test scores
Generate 10 test scores between 40 and 60, again using an index starting with September and ending with June. Find the mean of the scores and add the difference between the mean and 80 to each of the scores.

```python
g = np.random.default_rng(0)
s = Series(g.integers(40,61,10), index=months)
print('Grades mean:', s.mean())
s_diff = s.apply(lambda x: x + (80 - s.mean()))
#s_diff = s + (80 - s.mean())
print('New grades mean:', s_diff.mean())
s_diff
```
